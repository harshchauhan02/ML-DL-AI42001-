```python
import numpy as np
from sklearn import datasets
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import pandas as pd
```

```python
#making a car cost prediction model 
df = pd.DataFrame()

#generate random car horsepower
horsepower=[]
#generate random max speed
max_speed=[]
#generate a list of random car safety index
safety_index=[]
#generate a list of random car cost
car_cost=[]

for i in range(500):
    horsepower.append(np.random.randint(100,300))
    max_speed.append(np.random.randint(100,300))
    safety_index.append(np.random.randint(100,300))
    if(horsepower[i]>200 and max_speed[i]>200 and safety_index[i]>200):
        car_cost.append(3)
    elif(horsepower[i]>200 and max_speed[i]>200 and safety_index[i]<=200):
        car_cost.append(3)
    elif(horsepower[i]>200 and max_speed[i]<=200 and safety_index[i]>200):
        car_cost.append(2)
    elif(horsepower[i]>200 and max_speed[i]<=200 and safety_index[i]<=200):    
        car_cost.append(2)
    elif(horsepower[i]<=200 and max_speed[i]>200 and safety_index[i]>200):
        car_cost.append(1)
    elif(horsepower[i]<=200 and max_speed[i]>200 and safety_index[i]<=200):
        car_cost.append(2)
    elif(horsepower[i]<=200 and max_speed[i]<=200 and safety_index[i]>200):
        car_cost.append(1)
    elif(horsepower[i]<=200 and max_speed[i]<=200 and safety_index[i]<=200):
        car_cost.append(1)

        

```

```python
X=np.array([horsepower,max_speed,safety_index]).T
y=np.array(car_cost)
```

```python
df=pd.DataFrame(X,columns=['horsepower','max_speed','safety_index'])
# df.head()

df['target']=y
df.head()
```

```python
cmap = ListedColormap(['#FF0000','#00FF00','#0000FF'])
# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1234)

plt.figure()
plt.scatter(X[:,1],X[:,0], c=y, cmap=cmap, edgecolor='k', s=20)
plt.show()
```

```python
plt.figure()
plt.scatter(X[:,1],X[:,2], c=y, cmap=cmap, edgecolor='k', s=20)
plt.show()
```

```python
from ClassMaterial.KNN import KNN
clf = KNN(k=20)
clf.fit(X_train, y_train)
predictions = clf.predict(X_test)

print(predictions)
print(len(predictions))
print(predictions == y_test)

#getting the accuracy
acc = np.sum(predictions == y_test) / len(y_test)
print(acc)
```
